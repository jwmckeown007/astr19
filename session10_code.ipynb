# Session 10 Create Triangularly Random Histogram Using MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.triangular(0,0.5,1,1000)
```

```python
plt.figure(figsize=(10,6))
plt.hist(data, bins=100, edgecolor='black')

plt.title('Histogram Plot Session 10')
plt.xlabel('Awesome X Axis')
plt.ylabel('Awesome Y Axis')

plt.savefig('session10_histogram.pdf')
plt.show()
```
