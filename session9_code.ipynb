# Session 9 Create Uniformly Random Histogram Plot Using MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.uniform(0,1,1000)
```

```python
plt.figure(figsize=(10,6))
plt.hist(data, bins=100, edgecolor='black')

plt.title('Histogram Plot Session 9')
plt.xlabel('Awesome X Axis')
plt.ylabel('Awesome Y Axis')

plt.savefig('session9_histogram.pdf')
plt.show()
```
