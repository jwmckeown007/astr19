# Session 8 Create Multipanel Plot Using MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,100)
```

```python
def sin(x):
    return np.sin(x)
def cos(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,6))

axs[0].plot(x, sin(x))
axs[0].set_title('Sine of x')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

axs[1].plot(x, cos(x))
axs[1].set_title('Cosine of x')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

plt.savefig('session8_plot.pdf')
plt.show()
```
